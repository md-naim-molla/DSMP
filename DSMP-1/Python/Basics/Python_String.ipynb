```python
# Printing Star using Nested loop
n = int(input('Enter an integer: '))
for i in range(1,n+1):
  for j in range(1,i+1):
    print('*',end=' ')
  print('')
```

```python
lower = int(input('Enter lower range: '))
upper = int(input('Enter upper range: '))
prime_number = []
for i in range(lower,upper+1):
    for j in range(2,i):
        if i%j == 0:
            break
    else:
        prime_number.extend([i])
print(prime_number)
```

```python
# Count the frequency of a particular character in a provided string
s = input('Enter a string: ')
c = input('Enter character: ')
count = 0
for i in s:
    if i == c:
        count += 1
print(s)
print(f'The numbers of {c} in this string is: ',count)
```

```python
# write a program which can remove a particular character from a string
s = input('Enter a string: ')
c = input('Enter Char: ')
string = ''
for i in s:
    if i == c:
        continue
    string += i
print(s)
print(f'The string without {c} is: ',string)
```

```python
# Write a program than can check whether a given string is palindrome or not
name = input('Enter a name: ')
for i in range(len(name)//2):
    if name[i] == name[-i-1]:
        something = True
    else:
        something = False
        break
if something == True:
    print('Palindrom')
else:
    print('Not palindrom')
```
