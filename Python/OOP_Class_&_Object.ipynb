## Object-oriented programming
Object-Oriented Programming (OOP) is a programming paradigm (style of coding) based on the concept of objects instead of only functions.
Objects contain both data (attributes) and behavior (methods) to Perform operartions.
OOP organizes software design around objects rather than functions and logic, making code modular, reusable, maintainable, and scalable.
### Components of OOP
* Class
* Object
* Encapsulation
* Inheritance
* Abstraction
* Polymorphism

### Class
A class is a blueprint for creating objects.

Python has both built-in classes (like list, set, dict) and user-defined classes.

All data types in Python are actually classes, and the variables created from them are objects.

Example: list is a class, L = [1,2,3] → L is an object.

Mobile = Class

Samsung = Object

A class defines the rules and structure that its objects must follow.
### Object
* Object is an instance of class
### Constructor
A constructor is a special method inside a class.

In Python, the constructor is the __init__() method.

It runs automatically when an object is created—you don't need to call it manually.

If you want to create a variable of a class, you should declare it within the constructor and it starts with self

### Magic Method
__init__ is a magic method that is called automatically when an object is created.
The part of the code that you don’t want users to access directly (such as configuration-related code) is usually written inside the constructor, and it runs automatically when the object is created.
* You can create your own datatype using magic method
### Self
Only an object of a class can access all the variables, methods, and attributes of that class. This is the golden rule of OOP.
Therefore, when any method inside a class wants to access another method or variable of the same class, it uses self, because self represents the object itself.


```python
class Atm:
  def __init__(self):
    self.password = ''
    self.balance = 0
    self.menu()

  def menu(self):
    user_input = int(input('''
    Hi, How can i help You?
    1. Press 1 to create pin
    2. press 2 to change pin
    3. press 3 to check balance
    4. press 4 to withdraw
    5. press anything to exit
    '''))

    if user_input == 1:
      self.pin_creation()
    elif user_input == 2:
      self.change_pin()
    elif user_input == 3:
      self.check_balance()
    elif user_input == 4:
      self.withdraw()
    else:
      print('Invalid Syntax')
      exit()



  def pin_creation(self):
      user_pin = input('Enter Your pin: ')
      self.password = user_pin
      user_balance = int(input('Enter your Balance: '))
      self.balance = user_balance
      print('Pin Created Successful')
      self.menu()

  def change_pin(self):
    old_pin = input('Enter your old pin')
    if old_pin == self.password:
      new_pin = input('Enter new pin')
      self.password = new_pin
      print('Pin Change Succefull')
    else:
      print('Your password in incorrect')
    self.menu()

  def check_balance(self):
    old_pin = input('Enter your password')
    if old_pin == self.password:
      print(f'Your balance is {self.balance} taka')
    else:
      print('Incorrect password')
    self.menu()

  def withdraw(self):
    old_pin = input('enter your password: ')
    if old_pin == self.password:
      amount = int(input('Enter amount: '))
      if amount <= self.balance:
        self.balance = self.balance - amount
        print('Withdraw successful')
      else:
        print('Sorry! Insufficient Amount')
    else:
      print('Incorrect Password')
    self.menu()
```

```python
class Fraction:
  def __init__(self,x,y):
    self.num = x
    self.den = y

  def __str__(self):
    return '{}/{}'.format(self.num,self.den)

  def __add__(self,others):
    num = self.num*others.den + self.den*others.num
    den = self.den*others.den
    return '{}/{}'.format(num,den)

  def __sub__(self,others):
    num = self.num*others.den - self.den*others.num
    den = self.den*others.den
    return '{}/{}'.format(num,den)

```

```python
fr1 = Fraction(3,4)
fr2 = Fraction(1,4)
```

```python
print(fr1+fr2)
```

```python
print(fr1-fr2)
```
